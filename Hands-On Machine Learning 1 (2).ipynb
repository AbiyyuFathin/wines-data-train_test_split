```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

### Membaca file wine.csv dengan library pandas

```python
wines = pd.read_csv('C:/Users/abiyy/Downloads/wine.csv')
wines
```

### Menyimpan sub list points ke variabel x dan price ke variabel y

```python
x = wines.price
y = wines.points

#x = wines[pd.notnull(wines.points)]
#y = wines[pd.notnull(wines.price)]
```

### Melakukan split data ke dalam Training dan Testing Data

xTrain menyimpan data dari variabel x untuk training data, xTest menyimpan data dari variabel x untuk testing data, yTrain menyimpan data dari variabel y untuk training data, dan yTest menyimpan data dari variabel y untuk testing data.

Lalu, train_test_split adalah command dari SciKit Learn untuk melakukan pemisahan training data dan testing data. Adapun test_size digunakan untuk menetapkan pembagian tersebut menjadi 40% untuk testing data dan 60% untuk training data.

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.4, random_state = 0)
```

### Menampilkan hubungan dari Training data dari points dan price dalam bentuk Scatter Plot

```python
sns.scatterplot(xTrain, yTrain)
ax = plt.gca()
ax.set_title("Scatter of Training Data")
```

### Menampilkan hubungan dari Testing data dari points dan price dalam bentuk Scatter Plot

```python
sns.scatterplot(xTest, yTest)
ax = plt.gca()
ax.set_title("Scatter of Testing Data")
```
